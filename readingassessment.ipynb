```python
#Definir la función BuscarX
def xlookup(lookup_value, lookup_array, return_array, if_not_found:str = ''):
    match_value = return_array.loc[lookup_array == lookup_value]
    if match_value.empty:
        return f'"{lookup_value}" not found!' if if_not_found == '' else if_not_found

    else:
        return match_value.tolist()[0]
```

### Se revisa con 'response' cuál fue la regla escogida por el sujeto para cada elección, buscando en la carta seleccionada, cual tiene true or false.

```python
def get_matched_column(row):
    response = row['response']
    if response == 1:
        if row['matched_by_color1'] == 1:
            return 'matched_by_color1'
        elif row['matched_by_number1'] == 1:
            return 'matched_by_number1'
        elif row['matched_by_shape1'] == 1:
            return 'matched_by_shape1'
    elif response == 2:
        if row['matched_by_color2'] == 1:
            return 'matched_by_color2'
        elif row['matched_by_number2'] == 1:
            return 'matched_by_number2'
        elif row['matched_by_shape2'] == 1:
            return 'matched_by_shape2'
    elif response == 3:
        if row['matched_by_color3'] == 1:
            return 'matched_by_color3'
        elif row['matched_by_number3'] == 1:
            return 'matched_by_number3'
        elif row['matched_by_shape3'] == 1:
            return 'matched_by_shape3'
    elif response == 4:
        if row['matched_by_color4'] == 1:
            return 'matched_by_color4'
        elif row['matched_by_number4'] == 1:
            return 'matched_by_number4'
        elif row['matched_by_shape4'] == 1:
            return 'matched_by_shape4'
    return None

# Crear la nueva columna 'matched_column' aplicando la función get_matched_column a cada fila
wisconsindf['matched_column'] = wisconsindf.apply(get_matched_column, axis=1)

```

### Se limpia los valores de matched column y se dejan solo las columnas de nuestro interés

```python
# Función para limpiar la columna 'matched_column'
def clean_matched_column(value):
    if pd.isna(value):
        return value
    return value.replace('matched_by_', '').replace('1', '').replace('2', '').replace('3', '').replace('4', '')

# Aplicar la función a la columna 'matched_column'
wisconsindf['matched_column'] = wisconsindf['matched_column'].apply(clean_matched_column)
wisconsindf.rename(columns={'matched_column': 'selected_rule'}, inplace=True)
```
